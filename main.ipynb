```python
# convert video to audio

import moviepy.editor as mp

def toAudio(video,audio_folder):
    
    clip = mp.VideoFileClip(video)
    
    a=os.path.splitext(video)[0]
    video_filename=os.path.basename(os.path.normpath(a))
    # print(video)
    # print(audio_folder)
    
    clip.audio.write_audiofile(str(audio_folder)+"/audio_{}.wav".format(video_filename))
    
```

```python
import os
import moviepy.editor as mp

video_dir='/home/bit/kesav/codes/data_prep/videos'
audio_dir='/home/bit/kesav/codes/data_prep/audio'

def Convert_Video_To_Audio(video_dir):
    for folder in os.listdir(video_dir):
        path = os.path.join(video_dir, folder)
        # print(path)

        #creating audio folder
        audio_folder=os.path.join(audio_dir,folder)

        isExist = os.path.exists(audio_folder)

        if isExist:
            continue

        os.mkdir(audio_folder)
        
        for video_files in os.listdir(path):
            # print(video)
            video_path=os.path.join(path,video_files)
            # print(video_path)
            toAudio(video_path,audio_folder)
            
        
a=Convert_Video_To_Audio(video_dir)
```

```python
#finding corrsponding .srt file from folder
import os
from datetime import datetime
from pydub import AudioSegment

audio_dir='/home/bit/Documents/kesav/codes/data_prep/audio/Human rights in india/audio_files'
chAudio_dir='/home/bit/Documents/kesav/codes/data_prep/audio/Human rights in india/chunk_audio'
srt_dir='/home/bit/Documents/kesav/codes/data_prep/audio/Human rights in india/Srt_Files_1'

for a_file in os.listdir(audio_dir):
    try:
        s_file=os.path.join(srt_dir,a_file)
        s_file=s_file.replace('.wav','.srt')
        
        #creating chunk directory
        a_file_ext=a_file.replace('.wav','')
        chunk_path = os.path.join(chAudio_dir, a_file_ext)
        
        isExist = os.path.exists(chunk_path)
        # print(s_file)
        if isExist:
            # print(s_file)
            continue
            
        os.mkdir(chunk_path)

        with open(s_file,'r') as ifile:
            cnt = 1
            list_time=[]
            transcript=[]
            for line in ifile:
                if cnt==1:
                    time=next(ifile, '').strip()
                    
                    #getting trancrips
                    tran=next(ifile, '').strip()
                    n_line=next(ifile, '').strip()

                    if n_line=="":
                        transcript[time]=tran
                        continue
                    tran_new=tran+" "+n_line
                    transcript[time]=tran_new
               
                    
                    list_time.append(time.split('-->'))         
                    
                    cnt+=1

                else:
                    if line.strip()=="":
                        next(ifile, '').strip()
                        time=next(ifile, '').strip()
                        
                        #getting trancrips
                        tran=next(ifile, '').strip()
                        n_line=next(ifile, '').strip()

                        if n_line=="":
                            transcript[time]=tran
                            continue
                        tran_new=tran+" "+n_line
                        transcript[time]=tran_new
               
                    
                        # timeReplace=time.replace(',', ' ')
                        
                        list_time.append(time.split('-->'))
        
        print(transcript)
        del_index=[]
        #structuring timestamp and converting to milliseconds
        list_time=list_time[:-1]
        
        for i in range(len(list_time)):
                
                for j in range(len(list_time[i])):
                                        
                        if j==0:
                                list_time[i][j]=list_time[i][j][:12]
                                # print("hello")
                                #converting string to time format
                                if len(list_time[i][j])==12:
                                    dt=datetime.strptime(list_time[i][j], '%H:%M:%S,%f').time()        
                                    
                                    #converting to seconds
                                    hour=dt.hour*3600
                                    min=dt.minute*60
                                    sec=dt.second
                                    millisec=dt.microsecond/1000000

                                    #converting to milliseconds
                                    total=(hour+min+sec+millisec)*1000

                                    list_time[i][j]=total
                                else:
                                    del_index.append(i)  
                        else:
                                list_time[i][j]=list_time[i][j][1:13]
                                # print("else")
                                #converting string to time format
                                if len(list_time[i][j])==12:
                                    dt=datetime.strptime(list_time[i][j], '%H:%M:%S,%f').time()        

                                    #converting to seconds
                                    hour=dt.hour*3600
                                    min=dt.minute*60
                                    sec=dt.second
                                    millisec=dt.microsecond/1000000

                                    #converting to milliseconds
                                    total=(hour+min+sec+millisec)*1000

                                    list_time[i][j]=total

        time_new=[]
        for i in range(len(list_time)):
            flag=0
            for j in del_index:
                if i==j:
                    flag=1
                    break
            if flag==0:
                time_new.append(list_time[i])
        
        # splitting the audio file based on timestamps
            
        audio = AudioSegment.from_wav(os.path.join(audio_dir,a_file))
    
        for index,value in enumerate(time_new):

            if index == len(time_new):
                break

            start=int(value[0])
            end=int(value[1])
            
            audio_chunk=audio[start:end]
            audio_chunk.export(chunk_path+"/{}_chunk_{}.wav".format(a_file,(index+1)), format="wav")
        break    
    
    except Exception as e:
        log=open('/home/bit/Documents/kesav/codes/data_prep/audio/IT-Fundamentals/log.txt','w')
        log.write(str(a_file)+':')
        log.write(str(e))
        log.write("")
        log.close()

```
